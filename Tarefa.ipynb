# Primeira Tarefa

```python
#Importando as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Corrigindo o dataset

df = pd.read_csv('dataset_nicolas.csv')
df['idade'] = pd.to_numeric(df['idade'],errors = 'coerce')
df['horas_estudo'] = pd.to_numeric(df['horas_estudo'],errors = 'coerce')
df['nota'] = pd.to_numeric(df['nota'],errors = 'coerce')
df['faltas'] = pd.to_numeric(df['faltas'],errors = 'coerce')
df['tempo_transporte'] = pd.to_numeric(df['tempo_transporte'],errors = 'coerce')
df.info()
df = df.dropna(subset = ['idade','horas_estudo','nota','faltas','tempo_transporte'])
df.info()
df['grupo'] = df['grupo'].str.lower().str.strip()
df.head()
```

## Pergunta: pessoas com mais horas de estudo tem notas maiores?

```python
#relação entre tempo de transporte e nota

df['horas_estudo'] = df['horas_estudo'].apply(int).astype(int)

relação_horas_estudo_nota = df.groupby('horas_estudo')['nota'].mean()
```

```python
#Criando o grafico

plt.figure(figsize = (6,4))
relação_horas_estudo_nota.plot(kind = 'bar', color = 'salmon')
plt.title('Nota Média por Horas de Estudo')
plt.xlabel('Horas de Estudo')
plt.ylabel('nota')
plt.show()
```

- #### Correlação ou Causalidade: Isso se encaixa em correlação, pois, de acordo com o próprio gráfico, mesmo pessoas que estudaram 12 horas ainda tiraram notas baixas, enquanto outras que estudaram poucas horas obtiveram notas melhores ou iguais. Isso pode ser justificado pelo fato de que não adianta estudar 12 horas se essas horas não forem bem aproveitadas.


- #### Observação: algo notavel é que as médias mais baixas são de quem tem teve mais horas de estudo, comprovando ainda mais que as horas de estudo não tem relação com a nota

## Pergunta 2: Pessoas mais velhas faltam mais?

```python
# Relaçao entre idade e nota

relacao_idade_falta = df.groupby('idade')['faltas'].mean()
```

```python
#Criando gráfico

plt.figure(figsize = (6,4))
relacao_idade_falta.plot(kind = 'bar', color = 'skyblue')
plt.title('')
plt.ylabel('faltas')
plt.xlabel('idade')
plt.show()
```

- #### de acordo com o grafico e possivel perceber que a idade não tem relação com as faltas, pois a diferença entre 19 e 14 anos é muito pouco
