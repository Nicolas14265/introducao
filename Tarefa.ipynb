# Primeira Tarefa

```python
import pandas as pd
import matplotlib.pyplot as plt
import unicodedata
```

```python
df = pd.read_csv('dataset_nicolas.csv')
df['idade'] = pd.to_numeric(df['idade'],errors = 'coerce')
df['horas_estudo'] = pd.to_numeric(df['horas_estudo'],errors = 'coerce')
df['nota'] = pd.to_numeric(df['nota'],errors = 'coerce')
df['faltas'] = pd.to_numeric(df['faltas'],errors = 'coerce')
df['tempo_transporte'] = pd.to_numeric(df['tempo_transporte'],errors = 'coerce')
df = df.dropna(subset = ['idade','horas_estudo','nota','faltas','tempo_transporte'])
df['grupo'] = df['grupo'].str.lower().str.strip()

def remover_acentos(texto):
    if isinstance(texto,str):
        normalized = unicodedata.normalize('NFKD',texto)
        return normalized.encode('ascii','ignore').decote('utf-8')
    return texto
```

## Pergunta: pessoas com mais horas de estudo tem notas maiores

```python
#relação entre tempo de transporte e nota

df['horas_estudo'] = df['horas_estudo'].apply(int).astype(int)

relação_horas_estudo_nota = df.groupby('horas_estudo')['nota'].mean()
```

```python
plt.figure(figsize = (6,4))
relação_horas_estudo_nota.plot(kind = 'bar', color = 'lightcoral')
plt.title('Nota Média por Horas de Estudo')
plt.xlabel('Horas de Estudo')
plt.ylabel('nota')
plt.show()
```

#### Isso se encaixa na correlação, pois de acordo com o proprio grafico mesmo pessoas que estudaram 12 horas ainda tiraram notas baixas ou quem estudou poucas horas tiraram notas melhores ou iguais, e isso pode ser justificado pois nada adianta estudar 12 horas sendo essas horas mal aproveitada.


#### uma observação é que quem tirou notas altas foram os que tiveram menos horas de estudo
